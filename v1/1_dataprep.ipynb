```python
import pandas as pd

# 요청한 컬럼 목록(순서 유지)
requested_cols = [
    "blockNumber","timeStamp","hash","nonce","blockHash","from","contractAddress","to",
    "value","tokenName","tokenSymbol","tokenDecimal","transactionIndex","gas","gasPrice",
    "gasUsed","cumulativeGasUsed","input","confirmations"
]

# CSV 파일 경로
csv_path = "picaartmoney_transactions_full.csv"

# CSV 로드 (필요하면 dtype을 조정하세요)
df_full = pd.read_csv(csv_path, dtype=str)

# 존재하는 컬럼만 선택
available_cols = [c for c in requested_cols if c in df_full.columns]
missing_cols = [c for c in requested_cols if c not in df_full.columns]

df = df_full[available_cols].copy()

print("선택된 컬럼:", available_cols)
if missing_cols:
    print("누락된 컬럼:", missing_cols)
```

```python
# tokenDecimal을 이용해 value 보정(새 컬럼 `value_adjusted` 생성)
# 숫자형으로 변환 후 계산
import numpy as np

df["tokenDecimal"] = pd.to_numeric(df["tokenDecimal"], errors="coerce").fillna(0).astype(int)
df["value"] = pd.to_numeric(df["value"], errors="coerce").fillna(0)

# 10**tokenDecimal로 나눠 소수 위치를 복원
# tokenDecimal이 클 경우를 대비해 float으로 계산
df["value_adjusted"] = df["value"] / (10 ** df["tokenDecimal"]).astype(float)

# 확인
df[["value","tokenDecimal","value_adjusted"]].head()

```

```python
# NetworkX 그래프 생성: G (노드: addr, 엣지: net_to)
# - 노드 속성: address, cycles, dayvaloutlier, uniqedgesIn, centralitypagerank
# - 엣지 속성: netValue, netCount, earliestTimestamp

import networkx as nx

# 전제: df, value_adjusted, timeStamp 컬럼이 존재한다고 가정
# 안전하게 타입 변환
df["value_adjusted"] = pd.to_numeric(df.get("value_adjusted", df.get("value", 0)), errors="coerce").fillna(0).astype(float)
df["timeStamp"] = pd.to_numeric(df.get("timeStamp", 0), errors="coerce").fillna(0).astype(int)

# 날짜 컬럼(일 단위)
df["day"] = pd.to_datetime(df["timeStamp"], unit="s").dt.date

# 1) (from,to) 쌍으로 엣지 집계
edge_agg = df.groupby(["from","to"]).agg(
    netValue=("value_adjusted","sum"),
    netCount=("hash","count"),
    earliestTimestamp=("timeStamp","min")
).reset_index()

# 2) DiGraph 생성 및 엣지 추가
G = nx.DiGraph()

# 모든 주소(송신/수신)를 노드로 추가
addresses = pd.Index(df["from"].fillna("")).append(pd.Index(df["to"].fillna(""))).unique()
for addr in addresses:
    if addr == "":
        continue
    G.add_node(addr, address=addr)

# 집계된 엣지 추가 (속성 포함)
for _, r in edge_agg.iterrows():
    src = r["from"]
    dst = r["to"]
    # 안전성: 빈 문자열 또는 NaN 건너뜀
    if pd.isna(src) or pd.isna(dst) or src == "" or dst == "":
        continue
    G.add_edge(src, dst,
               netValue=float(r["netValue"]),
               netCount=int(r["netCount"]),
               earliestTimestamp=int(r["earliestTimestamp"]))

# 3) PageRank(centralitypagerank) 계산 (가중치는 netValue)
if G.number_of_nodes() > 0:
    try:
        pr = nx.pagerank(G, weight="netValue")
    except Exception:
        # fallback: unweighted
        pr = nx.pagerank(G)
else:
    pr = {}

nx.set_node_attributes(G, pr, name="centralitypagerank")

# 4) uniqedgesIn (유일한 수신 엣지 개수)
for n in G.nodes():
    uniq_in = len(list(G.predecessors(n)))
    G.nodes[n]["uniqedgesIn"] = int(uniq_in)

# 5) cycles: 강연결요소(SCC) 크기가 2 이상이면 순환 포함으로 판단
sccs = list(nx.strongly_connected_components(G))
scc_size_map = {}
for comp in sccs:
    size = len(comp)
    for node in comp:
        scc_size_map[node] = int(size) if size > 1 else 0
for n in G.nodes():
    G.nodes[n]["cycles"] = scc_size_map.get(n, 0)

# 6) dayvaloutlier: 송신자 기준 일별 합계의 z-score가 3 초과 여부
# (정의: 각 지갑의 일별 total value_adjusted를 구해 z-score 검정)
outlier_map = {}
from_groups = df.groupby(["from","day"]) ["value_adjusted"].sum()
for addr, grp in from_groups.groupby(level=0):
    vals = grp.values.astype(float)
    if len(vals) <= 1:
        outlier_map[addr] = False
        continue
    mean = vals.mean()
    std = vals.std(ddof=0)
    if std == 0:
        outlier_map[addr] = False
    else:
        zs = (vals - mean) / std
        outlier_map[addr] = bool((abs(zs) > 3).any())

for n in G.nodes():
    G.nodes[n]["dayvaloutlier"] = bool(outlier_map.get(n, False))

# 결과 확인(요약)
print(f"Graph G: nodes={G.number_of_nodes()}, edges={G.number_of_edges()}")

# G가 노트북에서 사용 가능하도록 출력
G

```

```python
# Graph G 저장 및 불러오기 (gpickle / pickle fallback)
import os
import pickle

save_path = "G_graph.gpickle"

# 저장
try:
    try:
        nx.write_gpickle(G, save_path)
        print(f"Saved G -> {save_path} (nx.write_gpickle)")
    except AttributeError:
        with open(save_path, "wb") as f:
            pickle.dump(G, f)
        print(f"Saved G -> {save_path} (pickle)")
except Exception as e:
    print("Save failed:", e)
```

```python

# 불러오기
try:
    try:
        G_loaded = nx.read_gpickle(save_path)
        print(f"Loaded G_loaded: nodes={G_loaded.number_of_nodes()}, edges={G_loaded.number_of_edges()} (nx.read_gpickle)")
    except AttributeError:
        with open(save_path, "rb") as f:
            G_loaded = pickle.load(f)
        print(f"Loaded G_loaded: nodes={G_loaded.number_of_nodes()}, edges={G_loaded.number_of_edges()} (pickle)")
except Exception as e:
    print("Load failed:", e)

# 이후 셀에서 G_loaded 사용 가능
G_loaded
```
